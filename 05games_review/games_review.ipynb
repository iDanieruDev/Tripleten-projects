# Descripción del Proyecto
Trabajamos para la tienda online Ice que vende videojuegos por todo el mundo. Las reseñas de usuarios y expertos, los géneros, las plataformas (por ejemplo, Xbox o PlayStation) y los datos históricos sobre las ventas de juegos están disponibles en fuentes abiertas. Debemos identificar patrones que determinen si un juego tiene éxito o no. Esto nos permitirá detectar proyectos prometedores y planificar campañas publicitarias.

El dataset contiene la abreviatura ESRB. The Entertainment Software Rating Board (la Junta de clasificación de software de entretenimiento) evalúa el contenido de un juego y asigna una clasificación de edad como Adolescente o Adulto.

Lo datos proporcionados se remontan al año 2016.

## Corrección de datos

### Corregimos los valores nulos en `Name`

```python
data.query('Name.isna()')
```

Hay solo dos filas con valores nulos en `Name`, correpondienes al año 1993 para un juego de la consola GEN. Estas dos filas no aportan en nada y son desprecibles en relación al conjutno de datos que disponemos. Se procederá a borrarlas.

```python
data.dropna(subset=['Name'], inplace=True)
data.info()
```

Al eliminar las dos filas nulas en `Name`, sin querelo, corregimos tambien la columna `Genre`, la cual ya no tiene datos nulos.

### Corregimos (si es posible) los valores nulos en `Year_of_Release`

```python
#Imprimimos una muestra de alguno datos nulos de esta columna
data.query('Year_of_Release.isna()').head(10)
```

```python
#Analizamos las consolas a las que pertenecen estos videojuegos
data.query('Year_of_Release.isna()').pivot_table(index='Platform',values='NA_sales', aggfunc='count')
```

```python
#Analizamos los géneros a las que pertenecen estos videojuegos
data.query('Year_of_Release.isna()').pivot_table(index='Genre',values='NA_sales', aggfunc='count')
```

En la columna `Year_of_Release` aparecen 269 filas (1.6% de los datos) con datos nulos. Con la información que disponemos no me es posible determinar el año de estos 269 juegos. Podría buscar uno a unos los juegos en internet, pero no me parece una alternativa práctica ni de valor para el analisis. De momento, me limitaré a establecer los valores nulos de esta columna, con el valor numérico "-1". Por último, transformaré estos datos de float a int.

```python
#Cuando uso np.nan luego no me deja convertir a int
data['Year_of_Release'].fillna(np.nan, inplace=True)
data.info()
```

### Corregimos los valores nulos en `Critic_Score`:

Como vimos anteriormente en el histograma de esta columna, todos los valores se encuentran en un rango de 0 a 100 (lo cual es correcto). Pero existe una gran cantidad de datos ausentes.

```python
print("\n")
aux=len(data.query('Critic_Score.isna()'))
aux2= aux/len(data)
print_mejorado(f"Los juegos sin calificación (valores ausentes) son {aux} ({aux2:.1%})",24)
print("\n")
```

Mas de la mitad de los juegos en nuestro DataFrame no poseen una calificación de la crítica. Esto es un porcentaje mas que significativo. Sin embargo, no podemos inferir (ni mucho menos borrar) esto datos, así que de momento, para estandarizar nuestro DataFrame, de momento,  le asignaré el valor -1 a los valores ausentes.

```python
#data['Critic_Score'].fillna(-1, inplace=True)
data['Critic_Score'].fillna(np.nan, inplace=True)
data.info()
```

### Corregimos los valores nulos en `User_Score` 

Como vimos anteriormente, esta columna es tipo objeto. Analicemos los valores unicos.

```python
data['User_Score'].unique()
```

Aquí vemos que la mayoría de los datos son numericos a excepción de "tba" (to be determinated) que es un string y de los valores ausentes (nan). Veamos a que porcentaje de nuestro de set de datos corresponden estos valores.

```python
print("\n")
aux=len(data.query('User_Score=="tbd"'))
aux2= aux/len(data)
print_mejorado(f"Los juegos con una calificación de usuarios pendiente (tbd) son {aux} ({aux2:.1%})",24)
aux=len(data.query('User_Score.isna()'))
aux2= aux/len(data)
print_mejorado(f"Los juegos sin calificación (valores ausentes) son {aux} ({aux2:.1%})",24)
print("\n")
```

Tanto las calificaciones por determinar (tbh) como las ausentes (nan) no se pueden determinar. Podriamos intentar asignarles el valor promedio total para no afectar el promedio de la muestra. Sin embargo, esto podria arrojar conclusiones equivocadas dependiendo de como se analice. De momento, a ambos valores le asignaré el valor -1 y luego de esto, transformaré la columna a float64.

```python
data['User_Score'].fillna(np.nan, inplace=True)
data['User_Score'].replace("tbd", np.nan, inplace=True)
data['User_Score'] = data['User_Score'].astype('float64')
data.info()
```

Por último, tras realizar las correcciones, nos aseguramos por última vez usando un histograma que los datos se encuentren en los valores esperados. Para esto no se considerará los valores ausentes (-1)

```python
#histo_des(data.query('User_Score!=-1')['User_Score'])
histo_des(data['User_Score'])
```

### Analisis de ingresos por plafaformas (2012-2016)


Observa cómo varían las ventas de una plataforma a otra. Elige las plataformas con las mayores ventas totales y construye una distribución basada en los datos de cada año. Busca las plataformas que solían ser populares pero que ahora no tienen ventas. ¿Cuánto tardan generalmente las nuevas plataformas en aparecer y las antiguas en desaparecer?

```python
#Enriquecemos nuestros DF agregando una nueva columna con los ingresos totales
data_actual['total_sales']=data_actual['NA_sales']+data_actual['EU_sales']+data_actual['JP_sales']+data_actual['Other_sales']
data_actual.info()
```

En base a lo anterior vemos que para nuestro periodo seleccionado, las plataformas mas relevantes en relación al número de ventas fueron "PS4", "PS3", "X360", "3DS", "XOne" y "WiiU".
```python
(    data_actual
    .query('Platform==@plataformas')
    .pivot_table(index=('Year_of_Release'),columns='Platform', values='total_sales', aggfunc="sum")
    .plot(kind="line",linestyle='-', marker='o',figsize=(15,10), xlabel='Año', ylabel='Ventas (milllones USD)')
)
```

En la gráfica anterior, vemos que en 2012, las plataformas mas populares eran PS3, X360, seguida de la portatil 3DS. Pero Luego al siguiente año aparecio la siguiente generación de consolas (PS4 y XONE) pasando a ser, con diferencia notable, las mas significativas. Finalmente, en 2016 las plataformas con mayor relevancia, y por consecuencia, mas rentables son **PS4, XOne y 3DS**,a su vez,PS3 y X360 luego del 2014 muestran un continuó descenso de ventas. ### Analisis de ganacias por género

A continuación agruparé por tipos de videojuegos (género) y veremos si existen diferencias entre los géneros mas y menos rentables.

```python
def rango_inf (df):
    total_sales=df
    Q1=total_sales.quantile(0.25)
    Q2=total_sales.quantile(0.5)
    Q3=total_sales.quantile(0.75)
    IQR = Q3-Q1    
    return (Q2-1.5*IQR)

def rango_sup (df):
    total_sales=df
    Q1=total_sales.quantile(0.25)
    Q2=total_sales.quantile(0.5)
    Q3=total_sales.quantile(0.75)
    IQR = Q3-Q1    
    return (Q2+1.5*IQR)

data_genero=(data_actual
    .pivot_table(index='Genre',values='total_sales', aggfunc=('sum','mean','median','max',rango_inf, rango_sup))
    .sort_values(by='sum', ascending=False)
    .rename(columns={'sum': 'ventas', 'mean': 'promedio', 'median': 'mediana'})
    .reindex(columns=['ventas','promedio', 'mediana','rango_inf','rango_sup','max'])
            )
data_genero
```

```python
data_genero.plot(kind="bar",y="ventas", figsize=(15,8))
```

```python
data_actual.boxplot(column=['total_sales'],by=['Genre'],figsize=(15,10))
```

Los géneros mas populares son **Action, Shooting, Role-Playing y Sports**, sin embargo, se deduce que si un juego es de un género en particular no necesariamente implica que un juego individualmente vaya a generar mas o menos ganancias solo por pertener a dicho género, es decir, **no se puede generalizar unicamente considerando el género al que pertenece el juego.**  

Esto se aprecia, por ejemplo, observando el género Action, que para el periodo analizado fue el que registró mas ganancias, pero que estas ganancias fue gracias a algunos juegos que sobresalieron de los demás (valores atípicos), cuando la gran mayoria de los juegos de acción registraron ingresos mas bajos.

### Verificación de las hipótesis

#### Calificaciones promedio entre Xbox One y PC

A continuaciones probaremos la hipótesis que **las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.**

Para esto, planteamos nuestras hipostesis nula y alterna y verificamos la hipotesis nula.  
-La hipotesis nula es que las calificaciones promedio de las plataformas Xbox One y PC son iguales  
-La hipotesis alterna es que las calificaiones promedio de las plataformas Xbox One y PC son diferentes.

Estamos ante la verificación de una **hipótesis sobre la igualdad de las medias de dos poblaciones**, para probar la hipótesis de que las medias de dos poblaciones estadísticas son iguales según las muestras tomadas de ellas, es necesario aplicar el método scipy.stats.ttest_ind.

Para el cálculo de esta hipotesis, usaré un valor de significación (alpha) de un 5%. Es decir, si el valor-p obtenido en el análisis estadístico es menor que 0.05, se considerará que el resultado es estadísticamente significativo y se rechazará la hipótesis nula en favor de la hipótesis alterna.

```python
#aux=data_actual.query('Platform=="XOne" and User_Score>-1')['User_Score'].mean()
aux=data_actual.query('Platform=="XOne"')['User_Score'].mean()
print_mejorado(f"El promedio de las calificaciones de XOne es: {aux:.2}",20)
#data_actual.query('Platform=="XOne" and User_Score>-1')['User_Score'].hist(bins=20)
data_actual.query('Platform=="XOne"')['User_Score'].hist(bins=20)
plt.axvline(x=aux, color='r')
```

```python
#aux=data_actual.query('Platform=="PC" and User_Score>-1')['User_Score'].mean()
aux=data_actual.query('Platform=="PC"')['User_Score'].mean()
print_mejorado(f"El promedio de las calificaciones de PC es: {aux:.2}",20)

#data_actual.query('Platform=="PC" and User_Score>-1')['User_Score'].hist(bins=20)
data_actual.query('Platform=="PC"')['User_Score'].hist(bins=20)
plt.axvline(x=aux, color='r')
```

Ambas distribuciones parecen tener un sesgo hacia la izquierda.

Definimos nuestros muestras:

```python
calificaciones_xone = data_actual.query('Platform=="XOne" and User_Score.notnull()')['User_Score']
calificaciones_pc = data_actual.query('Platform=="PC" and User_Score.notnull()')['User_Score']
```

Aplicamos el test de Shapiro para evaluar la normalidad de los datos.

```python
stat, p = st.shapiro(calificaciones_xone)
if p > 0.05: print_mejorado("Las calificaciones de XOne tienen una distribución normal",20)  
else: print_mejorado("Las calificaciones de XOne NO tienen una distribución normal",20)
    
stat, p = st.shapiro(calificaciones_pc)
if p > 0.05: print_mejorado("Las calificaciones de PC tienen una distribución normal",20)  
else: print_mejorado("Las calificaciones de PC NO tienen una distribución normal",20)
```

Aplicamos el test de Levene para evaluar sin las varianzas de nuestras muestras son distintas.

```python
# Test de Levene para verificar si las varianzas son iguales
stat, p = st.levene(calificaciones_xone, calificaciones_pc)

if p > 0.05:
    print_mejorado("Las varianzas son iguales",20)  
else:
    print_mejorado("Las varianzas son diferentes",20)
```

Procedemos a probar nuestras hipotesis. Dado que el test de Levene arrojó que nuestras varianzas son distintas, en <code>st.ttest_ind</code> usaré el atributo <code>equal_var=False</code>.

```python
# Prueba las hipótesis
#La hipotesis nula es que las calificaciones promedio de las plataformas Xbox One y PC son iguales
#La hipotesis alterna es que las calificaciones promedio de las plataformas Xbox One y PC son diferentes.

#calficaciones de la plataforma Xone (sin contar las calificaciones ausentes)
#calificaciones_xone = data_actual.query('Platform=="XOne" and User_Score>-1')['User_Score']
#calificaciones_xone = data_actual.query('Platform=="XOne"')['User_Score']

#calificaciones de la plataforma PC (sin contar las calificaciones ausentes)
#calificaciones_pc = data_actual.query('Platform=="PC" and User_Score>-1')['User_Score']
#calificaciones_pc = data_actual.query('Platform=="PC"')['User_Score']

#establecemos un nivel crítico de significación estadística (5%)
alpha = 0.05

# Probamos la hipótesis de que las medias de las ventas de los dos plataformas son iguales
#nota: asumimos que la varianza es igual para ambos grupos
results = st.ttest_ind(calificaciones_xone, calificaciones_pc,equal_var=False) 


print('p-value:', results.pvalue) 

if (results.pvalue < alpha): # tu código: compara los valores p obtenidos con el nivel de significación estadística
    print_mejorado("Rechazamos la hipótesis nula",20)
    print_mejorado("Las ventas promedio de las plataformas es diferente",24)
else:
    print_mejorado("No rechazamos la hipótesis nula",20)
    print_mejorado("Las calificaciones promedio de las plataformas es igual",24)
```

* En el 2016 en Norteamerica y Europa las plataformas mas populares eran PS4 y XOne.
* En Japón la plataforma mas popular en los últimos 5 años es 3DS.
* En 2016 en Japón las consolas mas populares despues de 3DS son PS4 y PSV.

*Géneros*
* En Norteamerica y Europa el género mas popular es Action, seguido por Shotter, luego Sports y Role-Playing.
* Para Norteamerica el quinto género mas popular es Misc.
* Para Europa el quinto género mas popular es Racing.
* En Japón el género Role-Playing es el mas popular, seguido del género Action.  

*Clasificacion ESRB*
* En Norteamerica y Europa los juegos populares son clasificación M (Mature). Luego E, T y E+10.
* En Japón los juegos populares son clasificación E (Everyone). Luego T, M y E+10.

**Hipotesis**
* Las calificaciones promedio de los usuarios para los juegos de las plataformas XOne y PC son similares.
* Las calificaciones promedio de los usuarios para los juegos de los géneros de Action y Sports son diferentes.
