Nuestro trabajo es analizar el comportamiento de los clientes y determinar qué tarifa de prepago genera más ingresos.

## Agregamos datos por usuario

Procedemos a calcular la cantidad de llamadas realizadas, la duración total de llamadas, los mensajes enviados y los datos consumidos por cada usuario en un mes.

```python
# Calculamos el número de llamadas hechas por cada usuario por mes y guardamos el resultado.

numero_llamadas=data_calls.pivot_table(index=('user_id','mes'),values='id',aggfunc=('count'))
numero_llamadas = numero_llamadas.rename(columns={'id': 'num_llamadas'})
numero_llamadas
```

```python
# Calculamos la cantidad de minutos que usó cada usuario por mes y guardamos el resultado.
minutos=data_calls.pivot_table(index=('user_id','mes'), values='duracion_redondeada',aggfunc=('sum'))
minutos
```

```python
# Calculamos el número de mensajes enviados por cada usuario por mes y guardamos el resultado.
numero_mensajes=data_messages.pivot_table(index=('user_id','mes'), values='id',aggfunc=('count'))
numero_mensajes = numero_mensajes.rename(columns={'id': 'num_mensajes'})
numero_mensajes
```

```python
# Calculamos el volumen del trafico de internet usado por cada usuario por mes y guarda el resultado.
gb_usados=data_internet.pivot_table(index=('user_id','mes'), values='mb_used',aggfunc=('sum'))
gb_usados
```

Combinamos todos los data frames antes construidos en uno solo llamaado "Consumo usuarios" el cual almacena el consumo general de cada usuario por mes.

```python
# Combinamos los datos para las llamadas, minutos, mensajes e internet con base en el user_id y el mes

#consumo_usuarios=numero_llamadas.join(minutos, on=['user_id', 'mes']).fillna(0)
#consumo_usuarios=consumo_usuarios.join(numero_mensajes, on=['user_id', 'mes'])
#consumo_usuarios=consumo_usuarios.join(gb_usados, on=['user_id', 'mes']).fillna(0)
#consumo_usuarios=consumo_usuarios.sort_values(by=['user_id', 'mes'])

consumo_usuarios=pd.concat([numero_llamadas, minutos, numero_mensajes, gb_usados], axis=1).fillna(0)
consumo_usuarios
```

```python
# Añadimos la información del plan

#Creo un dataframe auxiliar que solo contiene las columnas user_id, plan, mes_inicio y mes_fin
data_users_plan=data_users.loc[:,['user_id','plan','mes_inicio','mes_fin']]

#Inicializó una nueva columna llamada 'mes'
data_users_plan['mes']=''

#recorro el data frame por index
for index, row in data_users_plan.iterrows():
    
    #almaceno el mes_inicio, user_id y plan de la fila en caad loop    
    mes_inicio=int(data_users_plan['mes_inicio'].loc[index])
    user_id= data_users_plan['user_id'].loc[index]
    plan=data_users_plan['plan'].loc[index]

    #si es posible (si no es NaN) almaceno el mes_fin, si no se puede, lo definimos en 12
    try:mes_fin=int(data_users_plan['mes_fin'].loc[index])  #intentamos calcular el mes de final de contrato
    except: mes_fin=12    #si es nan no se puede calcular y se define 12

    #A nuestro dataframe auxiliar, agrego columnas por cada mes que el usuario tiene el plan
    #Por ejemplo, si el mes_inicio es 3 y mes_fin es 7 entonces agregamos los meses 4,5,6 y 7 con el plan correspondiente

    for i in range(mes_inicio+1,mes_fin+1):
        data_users_plan = data_users_plan.append({'user_id':user_id,'plan': plan,'mes': f'{i:02}'}, ignore_index=True)
            
    data_users_plan.at[index, 'mes'] = f'{mes_inicio:02}'

meses_plan=data_users_plan.pivot_table(index=('user_id','mes'),values=('plan'),aggfunc=('first'))
meses_plan.head(10)    
```

## Calculo de ingreso mensual por usuario

***Calculamos el ingreso mensual por usuario, en base a la información especificada arriba.***

```python
# Calculamos el ingreso mensual para cada usuario

#Construyo una función que calcule el ingreso segun el plan y el consumo mensual
def calculo_ingresos(df):
    plan=df['plan']
    llamadas=df['num_llamadas']
    minutos=df['duracion_redondeada']
    mensajes=df['num_mensajes']
    gb_usados=math.ceil(df['mb_used']/1024)

    ingresos=0
    
    if plan == 'surf':
        ingresos=20
        if minutos>500:
            ingresos=ingresos+(minutos-500)*0.03
        if mensajes>50:
            ingresos=ingresos+(mensajes-50)*0.03
        if gb_usados>15:
            ingresos=ingresos+(gb_usados-15)*10
        return ingresos
    
    if plan == 'ultimate':
        ingresos=70
        if minutos>3000:
            ingresos=ingresos+(minutos-3000)*0.01
        if mensajes>1000:
            ingresos=ingresos+(mensajes-1000)*0.01
        if gb_usados>30:
            ingresos=ingresos+(gb_usados-30)*7
        return ingresos
        
    if plan == 'cancelado':
        return 0

#Aplicamos nuestra función con el metodo apply y almacenamos el resultado en una nueva columna llamada 'ingresos'
consumo_usuarios['ingresos'] = consumo_usuarios.apply(calculo_ingresos, axis=1)

consumo_usuarios.head(100)
```

## Probamos las hipótesis estadísticas.

**Hipotesis de ingresos**  

Nuestro cliente nos solicita evaluar si es que existen diferencias entre los ingresos del plan Surf y los ingresos del plan Ultimate.

Para esto, planteamos nuestras hipostesis nula y alterna y verificamos la hipotesis nula.
-La hipotesis nula es que el ingreso promedio de los planes Ultimate y Surf es igual  
-La hipotesis alterna es que el ingreso promedio de los planes Ultimate y Surf es diferente

```python
# Prueba las hipótesis
#La hipotesis nula es que el ingreso promedio de los planes Ultimate y Surf es igual
#La hipotesis alterna es que el ingreso promedio de los planes Ultimate y Surf es diferente

# ingresos de los usuarios por mes (Surf)
ingresos_surf = consumo_usuarios2.query('plan=="surf"')['ingresos']

# ingresos de los usuarios por mes (Ultimate)
ingresos_ultimate  = consumo_usuarios2.query('plan=="ultimate"')['ingresos']

#establecemos un nivel crítico de significación estadística (5%)
alpha = 0.05

# Probamos la hipótesis de que las medias de los ingresos de los dos planes son iguales
#nota: asumimos que la varianza es igual para ambos grupos
results = st.ttest_ind(ingresos_surf, ingresos_ultimate) 


print('p-value:', results.pvalue) 

if (results.pvalue < alpha): # tu código: compara los valores p obtenidos con el nivel de significación estadística
    print_mejorado("Rechazamos la hipótesis nula",20)
    print_mejorado("El ingreso promedio mensual  de los planes es diferente",24)
else:
    print_mejorado("No rechazamos la hipótesis nula",20)
    print_mejorado("El ingreso promedio mensual  de los planes es igual",24)
```
