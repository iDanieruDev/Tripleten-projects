No todos los valores ausentes afectan a la investigación. Por ejemplo, los valores ausentes en la pista y artista no son cruciales. En este caso serán remplazamos con `unknown`.

Ahora eliminaré los duplicados implícitos en la columna `genre` (si es que lo hay). Por ejemplo, el nombre de un género se puede escribir de varias formas. Dichos errores también pueden afectar a resultado.

```python
# inspeccionando los nombres de géneros únicos
df.sort_values('genre',inplace=True) #inplace debe ponerse en True para que se actualice el DF

html_table = pd.DataFrame({'genre': df['genre'].unique()}).to_html(index=False)
display(HTML(html_table))
```

En la lista encontramos duplicados implícitos del género `hiphop`. Estos son nombres escritos incorrectamente o nombres alternativos para el mismo género.

En este caso, estos 3 géneros corresponden al genero `hiphop`:
* `hip`
* `hop`
* `hip-hop`

Para deshacernos de ellos facilmente, construiré la función `replace_wrong_genres()` con dos parámetros: 
* `wrong_genres=` — la lista de duplicados
* `correct_genre=` — la string con el valor correcto

La función coregirá los nombres en la columna `'genre'` de la tabla `df`, es decir, remplazará cada valor de la lista `wrong_genres` con el valor en `correct_genre`.

```python
# función para reemplazar duplicados implícitos
def replace_wrong_genres(wrong_genres, correct_genre): 
    df['genre'] = df['genre'].replace(wrong_genres, correct_genre) 
```

Llamamos la función `replace_wrong_genres()` para retirar los duplicados implícitos (`hip`, `hop` y `hip-hop`) y los reemplace por `hiphop`:

```python
# eliminando duplicados implícitos
duplicados = ['hip', 'hop','hip-hop'] # lista de nombres mal escritos
nombre_correcto = 'hiphop' # el nombre correcto
replace_wrong_genres(duplicados, nombre_correcto) 
```

Nos aseguramos que los nombres duplicados han sido eliminados. Imprimimos nuevamente la lista de valores únicos de la columna `'genre'`:

```python
# revisando en busca de duplicados implícitos
#Ordenamos
df.sort_values('genre',inplace=True)
# convierte los valores únicos en una tabla HTML
html_table = pd.DataFrame({'genre': df['genre'].unique()}).to_html(index=False)
display(HTML(html_table))
```

### Hipótesis 1: comparar el comportamiento del usuario en las dos ciudades <a id='activity'></a>

De acuerdo con la primera hipótesis, los usuarios de Springfield y Shelbyville escuchan música de forma distinta. Comprobaremos esto utilizando los datos de tres días de la semana: lunes, miércoles y viernes.

* Dividiré a los usuarios en grupos por ciudad.
* Veremos cuántas pistas reprodujo cada grupo el lunes, el miércoles y el viernes.

```python
# contando las pistas reproducidas en cada ciudad
data_por_ciudad=df.groupby('city')
print(data_por_ciudad['track'].count())
```

Springfield ha reproducido más pistas que Shelbyville. Pero eso no implica necesariamente que los ciudadanos de Springfield escuchen música más a menudo. **Esta ciudad es simplemente más grande y hay más usuarios.**

Ahora agrupamos los datos por día de la semana y observamos el número de pistas reproducidas el lunes, miércoles y viernes.


```python
# calculando las pistas reproducidas en cada uno de los tres días
data_por_dia=df.groupby('day')
print(data_por_dia['track'].count())
```

El miércoles fue el día más silencioso de todos. Pero si consideramos las dos ciudades por separado podríamos llegar a una conclusión diferente.

Una vez que ya agrupamos por ciudad y día, ahora escribiré una función que agrupará ambos.

Crearé la función `number_tracks()` para calcular el número de canciones reproducidas en un determinado día y ciudad. Esta tendrá dos parámetros:
* día de la semana
* nombre de la ciudad

La función utilizará una variable para almacenar las filas de la tabla original, donde:
  * el valor de la columna `'day'` es igual al parámetro de día
  * el valor de la columna `'city'` es igual al parámetro de ciudad

Aplicaré un filtrado consecutivo con indexación lógica, filtrando primero por `day` y luego por `city`.

Después, contaré los valores de la columna `'user_id'` en la tabla resultante y regresaré este valor.

```python
def number_tracks(day,city):
    #Se filtra por día
    track_list = df[df['day'] == day]
    
    #A la filtración anterior ahora se filtra por ciudad
    track_list = track_list[track_list['city'] == city]
    
    #Contamos
    track_list_count=track_list['user_id'].count()
    return track_list_count
```

Llamamos a `number_tracks()` seis veces, (3 días, 2 ciudades)

```python
# el número de canciones reproducidas en Springfield el lunes
print(number_tracks('Monday','Springfield'))
```

```python
# el número de canciones reproducidas en Shelbyville el lunes
print(number_tracks('Monday','Shelbyville'))
```

```python
# el número de canciones reproducidas en Springfield el miércoles
print(number_tracks('Wednesday','Springfield'))
```

```python
# el número de canciones reproducidas en Shelbyville el miércoles
print(number_tracks('Wednesday','Shelbyville'))
```

```python
# el número de canciones reproducidas en Springfield el viernes
print(number_tracks('Friday','Springfield'))
```

```python
# el número de canciones reproducidas en Shelbyville el viernes
print(number_tracks('Friday','Shelbyville'))
```

A continuación, crearé una nueva tabla con la siguiente informacion:
* Los nombres de las columnas son: `['city', 'monday', 'wednesday', 'friday']`
* Los datos son los 6 resultados conseguidos usando `number_tracks()`

```python
entradas =['city', 'monday', 'wednesday', 'friday']
data_musica =[['Springfield',
               number_tracks('Monday','Springfield'),
               number_tracks('Wednesday','Springfield'),
               number_tracks('Friday','Springfield')],
              ['Shelbyville',
               number_tracks('Monday','Shelbyville'),
               number_tracks('Wednesday','Shelbyville'),
               number_tracks('Friday','Shelbyville')]]

tabla_dias_ciudades = pd.DataFrame(data=data_musica, columns=entradas)
```

```python
# tabla con los resultados
tabla_dias_ciudades
```

**Conclusiones**

Los datos revelan las diferencias en el comportamiento de los usuarios:

- En Springfield, el número de canciones reproducidas alcanzan el punto máximo los lunes y viernes mientras que los miércoles hay un descenso de la actividad.
- En Shelbyville, al contario, los usuarios escuchan más música los miércoles. La actividad de los usuarios los lunes y viernes es menor.

Así que la primera hipótesis parece ser correcta.

### Hipótesis 2: música al principio y al final de la semana <a id='week'></a>

De acuerdo con la segunda hipótesis, los lunes por la mañana y los viernes por la noche los ciudadanos de Springfield escuchan géneros que difieren de aquellos que los usuarios de Shelbyville disfrutan.

De la tabla principal, obtendré dos tablas, una para Springfield y otra para Shelbyville.
* Para Springfield — `spr_general`
* Para Shelbyville — `shel_general`

```python
#Obteniendo la tabla Para Springfield (spr_general)
spr_general=df[df['city'] == 'Springfield']
spr_general.head(10)
```

```python
# obteniendo la tabla para Shelbyville (shel_general)
shel_general=df[df['city'] == 'Shelbyville']
shel_general.head(10)
```

Ahora escribiré la función genre_weekday() con cuatro parámetros de entrada:
* Una tabla para los datos (`df`)
* El día de la semana (`day`)
* Hora de inicio en formato 'hh:mm' (`time1`)
* Hora de terminoa en formato 'hh:mm' (`time2`)

La función devolverá información de los 15 géneros más populares de un día determinado en un período entre dos marcas de fecha y hora.

```python
def genre_weekday(df,day,time1,time2):

    # Filtramos por día
    genre_df = df[df['day'] == day]
    
    # Filramos filas con valores de hora menores a la hora de termino
    genre_df = genre_df[genre_df['time'] < time2]

    # Filramos filas con valores de hora mayores a la hora de inicio
    genre_df = genre_df[genre_df['time'] > time1]
    
    # Agrupamos por género y luego contamos    
    genre_df_grouped = genre_df.groupby('genre')['genre'].count()
    #Agrupo por 'genre', selecciono la columna 'genre' y luego cuento

    # Ordenaremos el resultado en orden descendente (los géneros más populares aparecerán primero)
    genre_df_sorted = genre_df_grouped.sort_values(ascending=False) 
   
    # devolveremos la Serie con los 15 géneros más populares en un día y periodo de tiempo determinado
    return genre_df_sorted[:15]
```

Comparamos los resultados de la función `genre_weekday()`para Springfield y Shelbyville para
- Lunes por la mañana (de 7 a 11)
- Viernes por la tarde (de 17:00 a 23:00)

```python
# llamando a la función para el lunes por la mañana en Springfield (utilizando la tabla de Springfield)
pd.DataFrame(genre_weekday(spr_general,'Monday','07:00','11:00'))
```

```python
# llamando a la función para el lunes por la mañana en Shelbyville (utilizando la tabla de Shelbyville)
pd.DataFrame(genre_weekday(shel_general,'Monday','07:00','11:00'))
```

```python
# llamando a la función para el viernes por la tarde en Springfield (utilizando la tabla de Springfield)
pd.DataFrame(genre_weekday(spr_general,'Friday','17:00','23:00'))
```

```python
# llamando a la función para el viernes por la tarde en Shelbyville (utilizando la tabla de Shelbyville)
pd.DataFrame(genre_weekday(shel_general,'Friday','17:00','23:00'))
```

**Conclusión**

Habiendo comparado los 15 géneros más populares del lunes por la mañana podemos concluir lo siguiente:

1. Los usuarios de Springfield y Shelbyville escuchan música similar. Los cinco géneros más populares son los mismos, solo rock y electrónica han intercambiado posiciones.

2. En Springfield el número de valores ausentes resultaron ser tan altos que el valor `'unknown'` llegó al décimo. Esto significa que los valores ausentes forman una parte considerable de los datos, lo que podría ser la base de la cuestión sobre la fiabilidad de nuestras conclusiones.

Para el viernes por la tarde, la situación es similar. Los géneros individuales varían algo pero, en general, los 15 más populares son parecidos en las dos ciudades.

De esta forma, la segunda hipótesis ha sido parcialmente demostrada:
* Los usuarios escuchan música similar al principio y al final de la semana.
* No hay una gran diferencia entre Springfield y Shelbyville. En ambas ciudades, el pop es el género más popular.

Sin embargo, el número de valores ausentes hace este resultado un tanto cuestionable. En Springfield, hay tantos que afectan a nuestros 15 más populares. De no faltarnos esos valores, las cosas podrían parecer diferentes.
